```python
import numpy as np
```

```python
array=np.array([1,2,3])
print(array)
```

```python
#Create a NumPy array from a list.

l1=[1,2,3]
array2=np.array(l1)
print(array2)
```

```python
#Generate an array of zeros, ones, or a specific constant in NumPy.
zeros=np.zeros((4,4))
print("zeros array is \n",zeros)
ones=np.ones((4,4))
print("ones array is \n",ones)
constantarray = np.full((3, 3), 7)  # 3x3 array filled with 7
print("constant array with specific constant \n",constantarray)


```

```python
#Find the maximum, minimum, and mean of a NumPy array.
arr1=np.array([1,2,3])
mean=np.mean(arr1)
print("min of arr1",arr1.min())
print("max of arr1",arr1.max())
print("mean of arr1",mean)


```

```python
#Stack two NumPy arrays vertically and horizontally.

array1 = np.array([[1, 2, 3], [4, 5, 6]])
array2 = np.array([[7, 8, 9], [10, 11, 12]])
vertical_stack = np.vstack((array1, array2))
horizontal_stack = np.hstack((array1, array2))
print("Vertical Stack:\n", vertical_stack)
print("\nHorizontal Stack:\n", horizontal_stack)
```

```python
#Find unique elements in a NumPy array.
arr = np.array([1, 2, 2, 3, 4, 4, 5, 6])
unique_elements = np.unique(arr)
print("Unique elements:", unique_elements)
```

```python
#Generate a random number matrix of size 5*5 using NumPy.
random_matrix = np.random.rand(5, 5)
print("Random 5x5 Matrix:\n", random_matrix)
```

```python
#Extract an element in the matrix.
matrix = np.random.randint(1, 101, (5, 5))  
print("Matrix:\n", matrix)
element = matrix[2, 3]
print("\nExtracted Element at (2,3):", element)
```

```python
#What is the relationship between confirmed cases and deaths? (Scatter plot)
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("country_wise_latest_modified.csv") 
plt.figure(figsize=(8,5))
sns.scatterplot(x=df["Confirmed"], y=df["Deaths"], alpha=0.6, color="red")
plt.xlabel("Confirmed Cases")
plt.ylabel("Deaths")
plt.title("Scatter Plot: Confirmed Cases vs Deaths")
plt.grid(True)
plt.show()

```

```python
#How does the number of active cases vary across different countries? (Box plot)
df = pd.read_csv("country_wise_latest_modified.csv")  # Load your dataset
plt.figure(figsize=(12,6))
sns.boxplot(x="Country/Region", y="Active", data=df)
plt.xticks(rotation=45)  
plt.xlabel("Country")
plt.ylabel("Active Cases")
plt.title("Distribution of Active COVID-19 Cases Across Different Countries")
plt.show()

```

```python
#How do the number of new cases compare across different WHO regions? (Bar chart or Pie chart)
df = pd.read_csv("country_wise_latest_modified.csv")  # Load your dataset
plt.figure(figsize=(12,6))
sns.barplot(x="New cases", y="WHO Region", data=df)
plt.xticks(rotation=45)  
plt.xlabel(" New cases")
plt.ylabel("WHO regions")
plt.title("Distribution of Active COVID-19 Cases Across Different Countries")
plt.show()

```

```python
df = pd.read_csv("country_wise_latest_modified.csv")  
region_cases = df.groupby("WHO Region")["New cases"].sum().reset_index()
plt.figure(figsize=(10,6))
sns.barplot(x="WHO Region", y="New cases", data=region_cases, palette="viridis")
plt.xlabel("WHO Region")
plt.ylabel("Total New Cases")
plt.title("Comparison of New COVID-19 Cases Across WHO Regions")
plt.xticks(rotation=45)

plt.show()
```
